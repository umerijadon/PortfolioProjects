# BMI Calculator

```python
name = input("Enter your name: ")

weight = int(input("Enter your weight in kilograms: "))

height = float(input("Enter your height in meters: "))

BMI = (weight)/(height * height)

print(BMI)

if BMI>0:
    if(BMI<18.5):
        print(name+", you're underweight. Eat up!")
    elif(BMI<=24.9):
        print(name+", you're normal weight. Keep it up!")
    elif(BMI<29.9):
        print(name+", you're overweight. Go exercise!")
    elif(BMI<30.0):
        print(name+", you're obese. Work on your diet!")
    else:
        print(name+", you're severely obese. Seek help..")
else:
    print("Try again!")
```
